## Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Reading Dataset

```python
dataset = pd.read_csv('machine.csv')
dataset
```

## Deleting unnecessary columns

```python
dataset_temp = dataset.drop('vendor name', axis = 1)
new_dataset = dataset_temp.drop('Model Name', axis = 1)
new_1_dataset = new_dataset.drop('PRP',axis = 1)
```

## Getting Data from columns of Dataset

```python
MYCT = new_1_dataset['MYCT'].values
MMIN = new_1_dataset['MMIN'].values
MMAX = new_1_dataset['MMAX'].values
CACH = new_1_dataset['CACH'].values
CHMIN = new_1_dataset['CHMIN'].values
CHMAX = new_1_dataset['CHMAX'].values
PRP = new_dataset['PRP'].values
```

## Printing Different Columns Values

```python
print(MYCT)
print(MMIN)
print(MMAX)
print(CACH)
print(CHMIN)
print(CHMAX)
print(PRP)
```

## Performing X^2, Y^2 and X*Y

```python
sqr_PRP = []
sqr_MYCT = []
sqr_MMIN = []
sqr_MMAX = []
sqr_CACH = []
sqr_CHMIN = []
sqr_CHMAX = []

prp_myct = []
prp_mmin = []
prp_mmax = []
prp_cach = []
prp_chmin = []
prp_chmax = []
for i in range(len(PRP)):
    sqr_PRP.append(PRP[i]*PRP[i])
    sqr_MYCT.append(MYCT[i]*MYCT[i])
    sqr_MMIN.append(MMIN[i]*MMIN[i])
    sqr_MMAX.append(MMAX[i]*MMAX[i])
    sqr_CACH.append(CACH[i]*CACH[i])
    sqr_CHMIN.append(CHMIN[i]*CHMIN[i])
    sqr_CHMAX.append(CHMAX[i]*CHMAX[i])
    prp_myct.append(MYCT[i]*PRP[i])
    prp_mmin.append(MMIN[i]*PRP[i])
    prp_mmax.append(MMAX[i]*PRP[i])
    prp_cach.append(CACH[i]*PRP[i])
    prp_chmin.append(CHMIN[i]*PRP[i])
    prp_chmax.append(CHMAX[i]*PRP[i])
```

## Creating new DataFrame 

```python
new_dataset['sqr_myct'] = sqr_MYCT
new_dataset['sqr_mmin'] = sqr_MMIN
new_dataset['sqr_mmax'] = sqr_MMAX
new_dataset['sqr_cach'] = sqr_CACH
new_dataset['sqr_chmin'] = sqr_CHMIN
new_dataset['sqr_chmax'] = sqr_CHMAX
new_dataset['prp_myct'] = prp_myct
new_dataset['prp_mmin'] = prp_mmin
new_dataset['prp_mmax'] = prp_mmax
new_dataset['prp_cach'] = prp_cach
new_dataset['prp_chmin'] = prp_chmin
new_dataset['prp_chmax'] = prp_chmax
new_dataset['sqr_prp'] = sqr_PRP

new_dataset
```

## Function for calculating Covariance

```python
def calCov(x,y):
    mean_x = sum(x)/209.0
    mean_y = sum(y)/209.0
    total_sum = 0
    for i in range(len(x)):
        total_sum += (x[i]-mean_x)*(y[i]-mean_y)
    return (total_sum / 208.0)
```

## Function for calculating variance

```python
def calVar(x):
    t_sum = 0
    m_x = sum(x)/209.0
    for i in range(len(x)):
        t_sum += pow((x[i]-m_x),2)
    return (t_sum/208.0)
```

## Calculating Beta1s

```python
w1 = calCov(MYCT,PRP)
z1 = calVar(MYCT)
beta_11 = w1 / z1

w2 = calCov(MMIN,PRP)
z2 = calVar(MMIN)
beta_12 = w2 / z2

w3 = calCov(MMAX,PRP)
z3 = calVar(MMAX)
beta_13 = w3 / z3

w4 = calCov(CACH,PRP)
z4 = calVar(CACH)
beta_14 = w4 / z4

w5 = calCov(CHMIN,PRP)
z5 = calVar(CHMIN)
beta_15 = w5 / z5

w6 = calCov(CHMAX,PRP)
z6 = calVar(CHMAX)
beta_16 = w6 / z6

print(beta_11,beta_12,beta_13,beta_14,beta_15,beta_16)
```

## Calculating Beta0s

```python
mean_prp = np.mean(PRP)
mean_myct = np.mean(MYCT)
mean_mmin = np.mean(MMIN)
mean_mmax = np.mean(MMAX)
mean_cach = np.mean(CACH)
mean_chmin = np.mean(CHMIN)
mean_chmax = np.mean(CHMAX)

beta_01 = mean_prp - (beta_11 * mean_myct)
beta_02 = mean_prp - (beta_12 * mean_mmin)
beta_03 = mean_prp - (beta_13 * mean_mmax)
beta_04 = mean_prp - (beta_14 * mean_cach)
beta_05 = mean_prp - (beta_15 * mean_chmin)
beta_06 = mean_prp - (beta_16 * mean_chmax)

print(beta_01,beta_02,beta_03,beta_04,beta_05,beta_06)
```

## Fitting Data

```python
prp_np = np.array(PRP)
myct_np = np.array(MYCT)
mmin_np = np.array(MMIN)
mmax_np = np.array(MMAX)
cach_np = np.array(CACH)
chmin_np = np.array(CHMIN)
chmax_np = np.array(CHMAX)

plt.scatter(myct_np,prp_np)
plt.plot(myct_np,beta_01+(beta_11*myct_np))
plt.title('MYCT vs PRP')
plt.xlabel('MYCT')
plt.ylabel('PRP')
plt.show()

plt.scatter(mmin_np,prp_np)
plt.plot(mmin_np,beta_02+(beta_12*mmin_np))
plt.title('MMIN vs PRP')
plt.xlabel('MMIN')
plt.ylabel('PRP')
plt.show()

plt.scatter(mmax_np,prp_np)
plt.plot(mmax_np,beta_03+(beta_13*mmax_np))
plt.title('MMAX vs PRP')
plt.xlabel('MMAX')
plt.ylabel('PRP')
plt.show()

plt.scatter(cach_np,prp_np)
plt.plot(cach_np,beta_04+(beta_14*cach_np))
plt.title('CACH vs PRP')
plt.xlabel('CACH')
plt.ylabel('PRP')
plt.show()

plt.scatter(chmin_np,prp_np)
plt.plot(chmin_np,beta_05+(beta_15*chmin_np))
plt.title('CHMIN vs PRP')
plt.xlabel('CHMIN')
plt.ylabel('PRP')
plt.show()

plt.scatter(chmax_np,prp_np)
plt.plot(chmax_np,beta_06+(beta_16*chmax_np))
plt.title('CHMAX vs PRP')
plt.xlabel('CHMAX')
plt.ylabel('PRP')
plt.show()
```

## Calculating Fitted Values

```python
fit_prp_myct = []
fit_prp_mmin = []
fit_prp_mmax = []
fit_prp_cach = []
fit_prp_chmin = []
fit_prp_chmax = []

for i in range(len(PRP)):
    val = beta_01 + beta_11 * MYCT[i]
    fit_prp_myct.append(val)
print(fit_prp_myct)

for i in range(len(PRP)):
    val = beta_02 + beta_12 * MMIN[i]
    fit_prp_mmin.append(val)
print(fit_prp_mmin)

for i in range(len(PRP)):
    val = beta_03 + beta_13 * MMAX[i]
    fit_prp_mmax.append(val)
print(fit_prp_mmax)

for i in range(len(PRP)):
    val = beta_04 + beta_14 * CACH[i]
    fit_prp_cach.append(val)
print(fit_prp_cach)

for i in range(len(PRP)):
    val = beta_05 + beta_15 * CHMIN[i]
    fit_prp_chmin.append(val)
print(fit_prp_chmin)

for i in range(len(PRP)):
    val = beta_06 + beta_16 * CHMAX[i]
    fit_prp_chmax.append(val)
print(fit_prp_chmax)
```

## Calculating Residuals

```python
residual_prp_myct = []
residual_prp_mmin = []
residual_prp_mmax = []
residual_prp_cach = []
residual_prp_chmin = []
residual_prp_chmax = []

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_myct[i]
    residual_prp_myct.append(res_i)
print(residual_prp_myct)

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_mmin[i]
    residual_prp_mmin.append(res_i)
print(residual_prp_mmin)

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_mmax[i]
    residual_prp_mmax.append(res_i)
print(residual_prp_mmax)

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_cach[i]
    residual_prp_cach.append(res_i)
print(residual_prp_cach)

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_chmin[i]
    residual_prp_chmin.append(res_i)
print(residual_prp_chmin)

for i in range(len(PRP)):
    res_i = PRP[i] - fit_prp_chmax[i]
    residual_prp_chmax.append(res_i)
print(residual_prp_chmax)
```

## Preparing Error Table Dataframe

```python
df = pd.DataFrame(list(zip(PRP,fit_prp_myct,residual_prp_myct,fit_prp_mmin,residual_prp_mmin,fit_prp_mmax,residual_prp_mmax,fit_prp_cach,residual_prp_cach,fit_prp_chmin,residual_prp_chmin,fit_prp_chmax,residual_prp_chmax)), columns = ['Actual PRP','fit_prp_myct','residual_prp_myct','fit_prp_mmin','residual_prp_mmin','fit_prp_mmax','residual_prp_mmax','fit_prp_cach','residual_prp_cach','fit_prp_chmin','residual_prp_chmin','fit_prp_chmax','residual_prp_chmax'])
df
```

## Calculating R^2

```python
cov_prp_myct_new = pow(calCov(PRP,MYCT),2)
var_myct_new = calVar(MYCT)
var_prp_new = calVar(PRP)
r_2_myct = cov_prp_myct_new / (var_myct_new * var_prp_new)
print("Coefficient of Determination(myct and prp):",r_2_myct)

cov_prp_mmin_new = pow(calCov(PRP,MMIN),2)
var_mmin_new = calVar(MMIN)
var_prp_new = calVar(PRP)
r_2_mmin = cov_prp_mmin_new / (var_mmin_new * var_prp_new)
print("Coefficient of Determination(mmin and prp):",r_2_mmin)

cov_prp_mmax_new = pow(calCov(PRP,MMAX),2)
var_mmax_new = calVar(MMAX)
var_prp_new = calVar(PRP)
r_2_mmax = cov_prp_mmax_new / (var_mmax_new * var_prp_new)
print("Coefficient of Determination(mmax and prp):",r_2_mmax)

cov_prp_cach_new = pow(calCov(PRP,CACH),2)
var_cach_new = calVar(CACH)
var_prp_new = calVar(PRP)
r_2_cach = cov_prp_cach_new / (var_cach_new * var_prp_new)
print("Coefficient of Determination(cach and prp):",r_2_cach)

cov_prp_chmin_new = pow(calCov(PRP,CHMIN),2)
var_chmin_new = calVar(CHMIN)
var_prp_new = calVar(PRP)
r_2_chmin = cov_prp_chmin_new / (var_chmin_new * var_prp_new)
print("Coefficient of Determination(chmin and prp):",r_2_chmin)

cov_prp_chmax_new = pow(calCov(PRP,CHMAX),2)
var_chmax_new = calVar(CHMAX)
var_prp_new = calVar(PRP)
r_2_chmax = cov_prp_chmax_new / (var_chmax_new * var_prp_new)
print("Coefficient of Determination(chmax and prp):",r_2_chmax)
```
